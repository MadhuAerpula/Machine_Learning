```python
""" When building a machine learning model, we split the dataset into:

    Training Set (70%) → Used to train the model.
    Testing Set (30%) → Used to evaluate model performance on unseen data.

We use train_test_split() from sklearn.model_selection to split the data."""

from sklearn.model_selection import train_test_split

X = df[['TV']]
Y = df['Sales']

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)

print("Training Feature Data (X_train):\n", X_train)
print("\nTesting Feature Data (X_test):\n", X_test)
print("\nTraining Target Data (y_train):\n", Y_train)
print("\nTesting Target Data (y_test):\n", Y_test)

"""
X → Features (Independent Variables).
y → Response Variable (Dependent Variable).
test_size=0.3 → 30% of the data is used for testing; the rest (70%) is used for training.
random_state=42 → Ensures the same random split every time for reproducibility."""
```

```python
X = df[['TV']]   # Independent Variable
y = df['Sales']  # Dependent Variable

# Split data into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Print the shape of training and testing sets
print("Shape of X_train (Feature Training Data):", X_train.shape)
print("Shape of X_test (Feature Testing Data):", X_test.shape)
print("Shape of y_train (Target Training Data):", y_train.shape)
print("Shape of y_test (Target Testing Data):", y_test.shape)
```

```python
X = df['TV']   # Independent Variable
y = df['Sales']  # Dependent Variable

# Split data into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Convert X_train and y_train to lists for manual calculations
X_train = list(X_train)
y_train = list(y_train)

# Calculate the necessary sums
N = len(X_train)
sum_X = sum(X_train)
sum_Y = sum(y_train)
sum_XY = sum([X_train[i] * y_train[i] for i in range(N)])
sum_X2 = sum([X_train[i] ** 2 for i in range(N)])

# Compute the slope (m) and intercept (c)
m = (N * sum_XY - sum_X * sum_Y) / (N * sum_X2 - sum_X ** 2)
c = (sum_Y - m * sum_X) / N

# Print coefficients
print("Calculated Coefficients:")
print("Slope (m):", m)
print("Intercept (c):", c)
```

```python
# Make predictions on the test set using the regression equation
y_pred = [m * X_test.iloc[i] + c for i in range(len(X_test))]

# Print actual vs predicted values
print("Actual vs Predicted Sales:")

for actual, predicted in zip(y_test, y_pred):
    print(f"Actual: {actual}, Predicted: {predicted}")
```

```python
from sklearn.linear_model import LinearRegression
X = df[['TV']]   # Independent Variable (Needs to be a 2D array)
y = df['Sales']  # Dependent Variable

# Split the data into 70% training and 30% testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Get the coefficients
m = model.coef_[0]  # Slope
c = model.intercept_  # Intercept

# Print coefficients
print("Linear Regression Model Coefficients:")
print("Slope (m):", m)
print("Intercept (c):", c)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Print actual vs predicted values
print("\nActual vs Predicted Sales:")
for actual, predicted in zip(y_test, y_pred):
    print(f"Actual: {actual}, Predicted: {predicted:.2f}")
```
