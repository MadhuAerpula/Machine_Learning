```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Week7_train.csv")
print(df)
```

```python
print(df.shape)
```

```python
# load 10% of the data from the actual df
sample_size = df.sample(frac=0.1, replace=False,random_state=1)
print(sample_size.shape)
```

```python
print(sample_size.shape)
```

```python
print(sample_size.head())
```

```python
print(sample_size['label'].unique())
print(sample_size['label'].value_counts())
print(sample_size['label'].value_counts().sum())
print(sample_size['label'].value_counts().sort_index())
```

```python
x = sample_size.drop('label', axis=1)
y = sample_size['label']
print(x.shape)
print(y.shape)
print(df.shape)
```
